# PHASE 1

**The census is conducted once in every 10 years hence census data is not available for each year for India. Since the data provided to us is limited within years 2001 and 2011, I will be collecting data such as state name, population data for each state, literacy rate and area of each state as per provided in census data of 2001 and 2011. In addition to that I will be collecting or webscrapping data for population density per sq.km for years 2001 and 2011.**

**I will be performing this task by either downloading excel files or by webscrapping using selenium method.**

**Importing libraries required to scrap data**

```python


import selenium
import pandas as pd
import time

from selenium import webdriver

from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException

import re
from selenium.webdriver.common.by import By

import warnings
warnings.filterwarnings('ignore')
```

```python
#connecting to driver
driver=webdriver.Chrome(r"D:\softwares\chromedriver_win32\chromedriver.exe")

#maximizing browser window
driver.maximize_window()
```

```python
#opening wikipedia.com website
url='https://en.wikipedia.org/wiki/2011_Census_of_India'
driver.get(url)
time.sleep(5)
```

```python
#scrapping data for state name
state=[]
try:
    description=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[2]/a')
    for i in description:
        state.append(i.text)
except NoSuchElementException:
    state.append('-')
```

```python
print(state)
len(state)
```

```python
#scrapping data for population statewise
population_2011=[]
try:
    description=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[5]')
    for i in description:
        population_2011.append(i.text)
except NoSuchElementException:
    population_2011.append('-')
```

```python
print(population_2011)
len(population_2011)
```

```python
#scrapping data for literacy rate statewise
lrate_2011=[]
try:
    description=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[10]')
    for i in description:
        lrate_2011.append(i.text)
except NoSuchElementException:
    lrate_2011.append('-')
```

```python
print(lrate_2011)
len(lrate_2011)
```

```python
#scrapping data for literacy rate statewise
area_2011=[]
try:
    description=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[13]')
    for i in description:
        area_2011.append(i.text)
except NoSuchElementException:
    area_2011.append('-')
```

```python
print(area_2011)
len(area_2011)
```

```python
#scrapping data for literacy rate statewise
popdensity_2011=[]
try:
    description=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[14]')
    for i in description:
        popdensity_2011.append(i.text)
except NoSuchElementException:
    popdensity_2011.append('-')
```

```python
print(popdensity_2011)
len(popdensity_2011)
```

```python
#creating dataframe for 2011 census scrapped data
data=pd.DataFrame({'state':state,'population':population_2011,'literacy_rate':lrate_2011,'area':area_2011,'density_pop':popdensity_2011})
data
```

```python
driver.close()
```

# Data cleaning
**removing commas population,area and population density columns**

```python
data['population']=data['population'].str.replace(',','')
data['area']=data['area'].str.replace(',','')
data['density_pop']=data['density_pop'].str.replace(',','')
```

```python
data
```

```python
#saving the 2011 census data into following named file

data.to_csv('2011_census_details')
```
